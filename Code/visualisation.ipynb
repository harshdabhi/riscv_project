```python
# Importing important library

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## AES Result Visualisation

```python
# Reading csv files
df_aes_baseline=pd.read_csv('aes_result/standard_results_aes.csv')
df_aes_accelerated=pd.read_csv('aes_result/accelerated_results_aes.csv')
```

```python
df_aes_baseline.columns
```

```python
sns.lineplot(df_aes_baseline,x='FileSize_KB',y='ExecutionTime_s',label='without B extension')
sns.lineplot(df_aes_accelerated,x='FileSize_KB',y='ExecutionTime_s',label='with B extension')
plt.ylabel('cpu time in ms')
```

```python
sns.lineplot(df_aes_baseline,x='FileSize_KB',y='Throughput_MBps',label='without B extension')
sns.lineplot(df_aes_accelerated,x='FileSize_KB',y='Throughput_MBps',label='with B extension')
plt.ylabel('Throughput_MBps')
```

## SHA Result Visualisation


```python
# Reading csv files
df_sha_baseline=pd.read_csv('sha_result/sha256_standard_results.csv')
df_sha_accelerated=pd.read_csv('sha_result/sha256_accelerated_results.csv')
```
